```python
#Imputing the missing values for goals and assists column with 0
sports_df['Goals'].fillna(0, inplace=True) 
sports_df['Assists'].fillna(0, inplace=True)  
```

```python
#Using the advanced imputation technique(knn) for filling in the missing values
from sklearn.impute import KNNImputer

#Imputing height column using age
cols = ['Age','Height']
knn = KNNImputer(n_neighbors=5)
knn.fit_transform(sports_df[cols])
df2 = pd.DataFrame(knn.transform(sports_df[cols]),columns=['Age', 'Height'])
sports_df['Height'] = df2['Height']
```

```python
#Imputing weight column using age
cols = ['Age','Weight']
knn = KNNImputer(n_neighbors=5)
knn.fit_transform(sports_df[cols])
df2 = pd.DataFrame(knn.transform(sports_df[cols]),columns=['Age', 'Weight'])
sports_df['Weight'] = df2['Weight']
```

```python
#Imputing passcompletionrate column using traininghours
cols = ['TrainingHours','PassCompletionRate']
knn = KNNImputer(n_neighbors=5)
knn.fit_transform(sports_df[cols])
df2 = pd.DataFrame(knn.transform(sports_df[cols]),columns=['TrainingHours','PassCompletionRate'])
sports_df['PassCompletionRate'] = df2['PassCompletionRate']
```

```python
#Imputing pressureperformanceimpact column using matchpressure
cols = ['MatchPressure','PressurePerformanceImpact']
knn = KNNImputer(n_neighbors=5)
knn.fit_transform(sports_df[cols])
df2 = pd.DataFrame(knn.transform(sports_df[cols]),columns=['MatchPressure','PressurePerformanceImpact'])
sports_df['PressurePerformanceImpact'] = df2['PressurePerformanceImpact']
```

```python
#Imputing pressureperformanceimpact column using matchpressure
cols = ['TrainingHours','EffectiveTraining']
knn = KNNImputer(n_neighbors=5)
knn.fit_transform(sports_df[cols])
df2 = pd.DataFrame(knn.transform(sports_df[cols]),columns=['TrainingHours','EffectiveTraining'])
sports_df['EffectiveTraining'] = df2['EffectiveTraining']
```

```python
#Checking the null values after imputing all the null values
sports_df.isnull().sum()
```

```python
#Retrieving the total number of duplicate values present in the dataset
duplicates = sports_df.duplicated().sum()
print("Duplicates present (count):",duplicates)
duplicates = sports_df[sports_df.duplicated(keep=False)]
display(duplicates)
```

```python
#Dropping the duplicate values with identical coulmns throught the row
sports_df.drop_duplicates(inplace=True)
sports_df
```

```python
#Creating a column "ID" with incrementing values for further analysis
cols=list(sports_df.columns)
sports_df['ID'] = range(1, len(sports_df) + 1)
sports_df=sports_df[['ID'] + cols]
sports_df
```

```python
#Identifying the number of outliers present in each numeric column using z-score statistical method
def outliers_IQR(df):
    q1=df.quantile(0.25)
    q3=df.quantile(0.75)
    IQR=q3-q1
    outliers = df[((df<(q1-1.5*IQR)) | (df>(q3+1.5*IQR)))]
    return outliers

numeric_cols = sports_df.select_dtypes(include=['float64', 'int64']).columns
sports_df[numeric_cols] = sports_df[numeric_cols].apply(pd.to_numeric, errors='coerce')
non_numeric_cols = sports_df.columns.difference(numeric_cols)
numeric_df = sports_df.drop(columns=non_numeric_cols)

for i in numeric_df.columns:
    outliers = outliers_IQR(sports_df[i])
    print('number of outliers in',i,': '+ str(len(outliers)))
```

```python
#Visualizing the columns with outliers using box plots
for i in numeric_df.columns:
    outliers = outliers_IQR(sports_df[i])
    print('Number of outliers in', i, ':', len(outliers))
    plt.figure(figsize=(8, 6))
    sns.boxplot(x=numeric_df[i])
    plt.title('Box plot for ' + i)
    plt.show()
```

```python
#Iterating through each row to find the rows with outliers in the dataframe
def outliers_IQR_row_wise(df):
    q1 = df.quantile(0.25)
    q3 = df.quantile(0.75)
    IQR = q3 - q1
    lower_bound = q1 - 1.5 * IQR
    upper_bound = q3 + 1.5 * IQR
    outliers_mask = ((df < lower_bound) | (df > upper_bound)).any(axis=1)
    return outliers_mask

outliers_mask = outliers_IQR_row_wise(numeric_df)
outliers_rows = numeric_df[outliers_mask]
print("Rows with outliers:")
outliers_rows
```

```python
def replace_outliers_with_mean(df):
    numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
    for col in numeric_cols:
        df[col] = pd.to_numeric(df[col], errors='coerce')
        original_dtype = df[col].dtype
        q1 = df[col].quantile(0.25)
        q3 = df[col].quantile(0.75)
        IQR = q3 - q1
        lower_bound = q1 - 1.5 * IQR
        upper_bound = q3 + 1.5 * IQR
        outliers_mask = ((df[col] < lower_bound) | (df[col] > upper_bound))
        df.loc[outliers_mask, col] = df[col].mean()
        if original_dtype == 'int64':
            df[col] = df[col].astype('int64')
    return df

sports_df = replace_outliers_with_mean(sports_df)
sports_df['Goals'] = sports_df['Goals'].astype(int)
print("DataFrame with outliers replaced by mean:")
sports_df
```

```python
#Creating a new csv file with the cleaned data for further tasks
sports_df.to_csv('Sports Dataset1.csv', index=False)
print("Cleaned sports dataset saved to Sports Dataset1.csv")
```
