```python
#Visualizing the distribution of the data using histograms before proceeding with appropriate imputation for each column
cols_to_visualize = ['Age', 'Height', 'Weight', 'Goals', 'Assists', 'PassCompletionRate','PressurePerformanceImpact','EffectiveTraining']
num_cols = len(cols_to_visualize)
num_rows = (num_cols + 3) // 4 
fig, axes = plt.subplots(nrows=num_rows, ncols=4, figsize=(15, num_rows * 3))
for i, col in enumerate(cols_to_visualize):
    row, col_index = divmod(i, 4)
    try:
        axes[row, col_index].hist(sports_df[col], bins='auto', edgecolor='black')
        axes[row, col_index].set_title(f'Distribution of {col}')
        axes[row, col_index].set_xlabel(col)
        axes[row, col_index].set_ylabel('Frequency')
    except IndexError:
        axes[row, col_index].axis('off')
fig.suptitle('Distribution of Columns that contains null values in Sports Data',fontsize=19)
plt.tight_layout()
plt.show()
```

```python
#Imputing the missing values for goals and assists column with 0
sports_df['Goals'].fillna(0, inplace=True) 
sports_df['Assists'].fillna(0, inplace=True)  
```

```python
#Using the advanced imputation technique(knn) for filling in the missing values
sports_knn = sports_df.copy(deep=True)
from sklearn.impute import KNNImputer
sports_knn = sports_df.copy(deep=True)
cols = ['Height','PassCompletionRate','PressurePerformanceImpact','EffectiveTraining']
knn_imputer = KNNImputer(n_neighbors=2, weights="uniform")
sports_knn[cols] = knn_imputer.fit_transform(sports_knn[cols])
```

```python
#Using predictive modelling to fill in the weight column with null values to achieve more accuracy and to avoid any potential bias

from sklearn.linear_model import LinearRegression
# Spliting the dataset into two parts,one with missing age values and one without
missing_weight_df = sports_knn[sports_knn['Weight'].isna()]
filled_weight_df = sports_knn.dropna(subset=['Weight'])
features = ['Height','Age']  
#Training a regression model on rows without missing weight values
X_train = filled_weight_df[features]
y_train = filled_weight_df['Weight']
model = LinearRegression()
model.fit(X_train, y_train)
#Predicting weight for rows with missing age values
X_missing = missing_weight_df[features]
missing_weight_predictions = model.predict(X_missing)
#Filling in missing age values in the original DataFrame
sports_knn.loc[sports_knn['Weight'].isna(),'Weight'] = missing_weight_predictions
sports_knn
```

```python
#Checking the null values after imputing all the null values
sports_knn.isnull().sum()
```

```python
#Retrieving the total number of duplicate values present in the dataset
duplicates = sports_knn.duplicated().sum()
print("Duplicates present (count):",duplicates)
duplicates = sports_knn[sports_knn.duplicated(keep=False)]
display(duplicates)
```

```python
#Dropping the duplicate values with identical coulmns throught the row
sports_knn.drop_duplicates(inplace=True)
sports_knn
```

```python
cols=list(sports_knn.columns)
sports_knn['Index'] = range(1, len(sports_knn) + 1)
sports_df=sports_knn[['Index'] + cols]
sports_df
```

```python
#Identifying the outliers using z-score statistical method
import numpy as np
def detect_outliers(df, cols):
    means = {}
    stds = {}
    outliers = {}
    for col in cols:
        means[col] = np.mean(df[col])
        stds[col] = np.std(df[col])
        threshold = 3  
        z_scores = (df[col] - means[col]) / stds[col]
        outliers[col] = df[(z_scores < -threshold) | (z_scores > threshold)]
    return means, stds, outliers
cols_to_check = ['Age', 'Height', 'Weight', 'Goals', 'Assists', 'YellowCards', 'RedCards', 
                 'PassCompletionRate', 'DistanceCovered', 'Sprints', 'ShotsOnTarget', 'TacklesWon',
                 'CleanSheets', 'TrainingHours', 'MatchPressure'] 
means, stds, outliers_by_col = detect_outliers(sports_df.copy(), cols_to_check)
print("\nOutliers by Column:")
for col, outlier_df in outliers_by_col.items():
    print(f"\n{col} Outliers (count): {len(outlier_df)}")
```

```python
#Itering through each row to find the rows with outliers in the dataframe
for col, outlier_df in outliers_by_col.items():
    for index, row in outlier_df.iterrows():
        print(f"Row with outlier in '{col}':{row}")
```

```python
#Visualizing the outliers using box plots
def detect_outliers_zscore(df, col, threshold=3):
    means = df[col].mean()
    stds = df[col].std()
    z_scores = (df[col] - means) / stds
    outliers = df[(z_scores < -threshold) | (z_scores > threshold)]
    return outliers
cols_with_outliers = ['Height', 'Weight', 'Goals'] 
fig, axes = plt.subplots(nrows=len(cols_with_outliers), figsize=(5, 15))  
for i, col in enumerate(cols_with_outliers):
    outliers_df = detect_outliers_zscore(sports_knn.copy(), col) 
    axes[i].boxplot([sports_df[col], outliers_df[col]], vert=True, patch_artist=True, notch=True, labels=['All Data', 'Outliers'])
    axes[i].set_title(f"{col} Distribution (with Outliers)")
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Value')
fig.suptitle('Boxplots for Columns with Outliers (Z-score)', fontsize=12)
plt.tight_layout()
plt.show()
```
